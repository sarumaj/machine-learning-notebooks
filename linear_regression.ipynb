```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# load dataset
data = load_diabetes()
y = data.target.reshape(-1, 1)
x = data.data[:, 2].reshape(-1, 1)
labels = data.feature_names
print(labels)
_ = plt.scatter(x, y)
_ = plt.title('diabetes progression vs bmi')
_ = plt.xlabel('bmi')
_ = plt.ylabel('diabetes progression')
```

```python
reg = LinearRegression().fit(x, y)
print("coefficients: ", reg.coef_, "intercept: ", reg.intercept_)
y_pred = reg.predict(x)
_ = plt.scatter(x, y)
_ = plt.plot(x, y_pred, color='red')
_ = plt.title('diabetes progression vs bmi, linear regression,\nR2: %.2f, RMS: %.2f' % (r2_score(y, y_pred), mean_squared_error(y, y_pred)))
_ = plt.xlabel('bmi')
_ = plt.ylabel('diabetes progression')
```
