```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
```

```python
np.random.seed(42)
x = np.random.normal(size=150)
y = (x > 0).astype(float)
x[x > 0] *= 4
x += .3 * np.random.normal(size=150)
x = x.reshape(-1, 1)
_ = plt.scatter(x, y)
_ = plt.title('Binary classification')
_ = plt.xlabel('x')
_ = plt.ylabel('y')
```

```python
reg = LogisticRegression().fit(x, y.ravel())
y_pred = reg.predict(x)
x_ordered = np.sort(x, axis=0)
y_pred_proba = reg.predict_proba(x_ordered)
_ = plt.scatter(x.ravel(), y)
_ = plt.plot(x_ordered, y_pred_proba[:, 1], color='red')
_ = plt.title(
    'Logistic regression, R2: %.2f, MSE: %.2f' %
    (r2_score(y, y_pred), mean_squared_error(y, y_pred))
)
_ = plt.xlabel('x')
_ = plt.ylabel('y')
```
