```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture
```

```python
# load the iris dataset
iris = load_iris()
X = iris.data[:, :2]  # we only take the first two features.
df = pd.DataFrame(X, columns=iris.feature_names[:2])
```

```python
_ = plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'])
_ = plt.title('Iris dataset')
_ = plt.xlabel('sepal length (cm)')
_ = plt.ylabel('sepal width (cm)')
```

```python
# fit a Gaussian Mixture Model with three components
random_state = np.random.RandomState(0)
gmm = GaussianMixture(n_components=3, random_state=random_state).fit(df)
```

```python
# predict the cluster for each data point
labels = gmm.predict(df)
```

```python
# check if model has converged
print(gmm.converged_)
```

```python
# print the number of iterations
print(gmm.n_iter_)
```

```python
#  get the final means
means = gmm.means_
print(means)
```

```python
#  get the final covariances
covariances = gmm.covariances_
print(covariances)
```

```python
# plot the data with clusters
_ = plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=labels)
_ = plt.title('Iris dataset clustered')
_ = plt.xlabel('sepal length (cm)')
_ = plt.ylabel('sepal width (cm)')
```
